linalg.solve for the solution of one linear system

```python
A = np.random.rand(3,3)
b = np.array([1,2,3])
x = np.linalg.solve(A,b)
print(x)
print(type(x))
```

```python
print('type of A =',type(A))
print('residual =', np.dot(A,x)-b) 
```

```python
# remember: the array vector does not are row vectors or column vectors
y = x.T
print(y.shape)

print('A*x = ',A*x)
print('A*y-A*x =', A*y-A*x)
print(' A = ', A)
print(' x = ', x)
print('A[:,0]*x[0] = ', A[:,0]*x[0])
print('A[:,1]*x[1] =', A[:,1]*x[1])
print( 'A[:,2]*x[2] =',A[:,2]*x[2])
```

Gaussian elimination in floating point arithmetic

```python
import numpy as np
ex = 2
par = 1.0e-17 # 10^(-17)
if ex==1:
# Example 1 well conditioned
 A=np.array( [[par,1.],[1., 2.]])
 b= np.array([[1.],[3.]])
else:
# Example 2 ill conditioned, but a scaling could solve the problem
 A=np.array( [[10.,1./par],[1., 2.]])
 b= np.array([[1./par],[3.]])


print(A)
print(b)
```

```python
Ah = np.hstack((A,b))
print(Ah)
```

```python
Ah[1,:]=Ah[1,:]-Ah[0,:]*(Ah[1,0]/Ah[0,0])
print(Ah)
```

solve the upper triangular system

```python
xg = np.zeros((2,1))
xg[1]= Ah[1,2]/Ah[1,1]
xg[0]= (Ah[0,2]-Ah[0,1]*xg[1])/Ah[0,0]
print('Solution after Gaussian Elimination \n',xg)
```

```python
x = np.linalg.solve(A,b)
print('linalg solve algorithm \n',x)
print("condition number",np.linalg.cond(A))
```

Solution of a lower triangular linear system:

```python
#definiamo la function
import numpy as np

def solL(L,b):
    """
    Solution of a lower triangular system,
    implementation by row 
    INPUT  =matrix L, known term b
    OUTPUT =solution x
    
    """
    (n,m) = np.shape(b)
    x = np.zeros((n,1))
    x[0]=b[0]/L[0,0]
    for i in range(1,n):
        x[i]=(b[i] - np.dot(L[i,0:i],x[0:i]))/L[i,i]
    return x

n=100
L  = np.tril(np.random.rand(n,n))+np.eye(n)*1
xt = np.ones((n,1))
b  = np.dot(L,xt)
x  = solL(L,b)
print('absolute error')
print( np.max(abs(x-xt)))
print("determinant",np.prod(np.diag(L)) )
print("condition number",np.linalg.cond(L))
```
